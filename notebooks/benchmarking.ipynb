```python
import os, json, re, unicodedata, glob
from pathlib import Path
from datasets import load_dataset, Audio
import soundfile as sf
import librosa
import numpy as np
from tqdm.auto import tqdm

ROOT = Path("/kaggle/working/asr_vi_eval")
WAV_DIR = ROOT / "audio"
MAN_DIR = ROOT / "manifests"
WAV_DIR.mkdir(parents=True, exist_ok=True)
MAN_DIR.mkdir(parents=True, exist_ok=True)

def normalize_vi(s):
    s = unicodedata.normalize("NFC", str(s).lower())
    s = re.sub(r"[^\w\sàáảãạăằắẳẵặâầấẩẫậèéẻẽẹêềếểễệìíỉĩịòóỏõọôồốổỗộơờớởỡợùúủũụưừứửữựỳýỷỹỵđ]", " ", s)
    s = re.sub(r"\s+", " ", s).strip()
    return s

def write_wav(arr, sr, out):
    arr = np.asarray(arr)
    if arr.ndim > 1:
        arr = arr.mean(axis=1)
    if sr != 16000:
        arr = librosa.resample(arr.astype(np.float32), orig_sr=sr, target_sr=16000)
    sf.write(str(out), arr, 16000)
    return len(arr) / 16000

def get_audio(row):
    a = row["audio"]

    # datasets bản cũ: {"array": ..., "sampling_rate": ...}
    if isinstance(a, dict):
        if "array" in a and "sampling_rate" in a:
            return a["array"], a["sampling_rate"]
        if "path" in a and a["path"]:
            arr, sr = sf.read(a["path"])
            return arr, sr

    # datasets bản mới: torchcodec AudioDecoder
    if hasattr(a, "get_all_samples"):
        samples = a.get_all_samples()
        arr = samples.data
        sr = samples.sample_rate
        if hasattr(arr, "detach"):
            arr = arr.detach().cpu().numpy()
        return arr, sr

    raise ValueError(f"Unsupported audio type: {type(a)}")


def write_wav(arr, sr, out):
    arr = np.asarray(arr)

    # torchcodec thường trả shape [channels, samples]
    if arr.ndim == 2:
        if arr.shape[0] <= 8 and arr.shape[1] > arr.shape[0]:
            arr = arr.mean(axis=0)
        else:
            arr = arr.mean(axis=1)

    arr = arr.astype(np.float32)

    if sr != 16000:
        arr = librosa.resample(arr, orig_sr=sr, target_sr=16000)

    sf.write(str(out), arr, 16000)
    return len(arr) / 16000

def build_from_hf(name, repo_id, split, text_col, config=None, limit=None):
    print("BUILD", name, repo_id, split)
    ds = load_dataset(repo_id, config, split=split) if config else load_dataset(repo_id, split=split)
    ds = ds.cast_column("audio", Audio(decode=True))
    out_dir = WAV_DIR / name
    out_dir.mkdir(parents=True, exist_ok=True)
    rows = []
    n = len(ds) if limit is None else min(limit, len(ds))
    for i in tqdm(range(n), desc=name):
        r = ds[i]
        text = normalize_vi(r.get(text_col, ""))
        if not text:
            continue
        arr, sr = get_audio(r)
        wav = out_dir / f"{i:07d}.wav"
        dur = write_wav(arr, sr, wav)
        if 0.3 <= dur <= 30:
            rows.append({"audio_filepath": str(wav), "duration": round(dur, 3), "text": text})
    man = MAN_DIR / f"{name}.test.jsonl"
    man.write_text("\n".join(json.dumps(x, ensure_ascii=False) for x in rows) + "\n", encoding="utf-8")
    print("WROTE", man, len(rows))
    return man

def build_carved_from_hf(name, repo_id, source_split, text_col, config=None, n_test=1000):
    print("CARVE", name, repo_id, source_split)
    ds = load_dataset(repo_id, config, split=source_split) if config else load_dataset(repo_id, split=source_split)
    ds = ds.cast_column("audio", Audio(decode=True))

    # chọn deterministic 1000 mẫu cuối theo index nếu không có path ổn định trong HF viewer
    idxs = list(range(len(ds)))[-n_test:]
    out_dir = WAV_DIR / name
    out_dir.mkdir(parents=True, exist_ok=True)
    rows = []
    for j, i in enumerate(tqdm(idxs, desc=name)):
        r = ds[i]
        text = normalize_vi(r.get(text_col, ""))
        if not text:
            continue
        arr, sr = get_audio(r)
        wav = out_dir / f"{i:07d}.wav"
        dur = write_wav(arr, sr, wav)
        if 0.3 <= dur <= 30:
            rows.append({"audio_filepath": str(wav), "duration": round(dur, 3), "text": text})
    man = MAN_DIR / f"{name}.test.jsonl"
    man.write_text("\n".join(json.dumps(x, ensure_ascii=False) for x in rows) + "\n", encoding="utf-8")
    print("WROTE", man, len(rows))
    return man
```

```python
# Official test
build_from_hf("vivos", "ademax/vivos-vie-speech2text", "test", "transcription")
build_from_hf("common_voice_vi", "tsdocode/common_voice_13_0_vi_pseudo_labelled", "test", "sentence")
build_from_hf("fleurs_vi", "google/fleurs", "test", "transcription", config="vi_vn")

# Carve 1000 từ train/source
build_carved_from_hf("fosd", "doof-ferb/fpt_fosd", "train", "transcription", n_test=1000)
build_carved_from_hf("vlsp2020_100h", "doof-ferb/vlsp2020_vinai_100h", "train", "transcription", n_test=1000)
build_carved_from_hf("lsvsc", "doof-ferb/LSVSC", "train", "transcription", n_test=1000)
build_carved_from_hf("vietsuperspeech", "thanhnew2001/VietSuperSpeech", "train", "text", n_test=1000)

# Nặng/gated, chạy sau nếu có quyền và đủ disk
# build_carved_from_hf("bud500", "linhtran92/viet_bud500", "train", "transcription", n_test=1000)
# build_from_hf("vietmed", "leduckhai/VietMed", "test", "text")
```
